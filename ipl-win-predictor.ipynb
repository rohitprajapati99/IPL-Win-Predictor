# IPL Win Predictor

### Import necessary libraries

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

### Load the dataset

```python
match_data = pd.read_csv("Desktop/IPL Win Predictor/match_data.csv",low_memory=False)
match_info = pd.read_csv("Desktop/IPL Win Predictor/match_info_data.csv",low_memory=False)
```

```python
match_info.head()
```

```python
match_data.head()
```

### Rename column to merge both datasets

```python
match_info.rename(columns={"id": "match_id"}, inplace=True)
df = match_data.merge(match_info,on="match_id")
```

```python
df.columns
```

```python
df['total_runs'] = df['runs_off_bat'] + df['extras']
```

```python
df.head()
```

```python
df.shape
```

### Adding all runs of each inning

```python
df.groupby(['match_id','innings']).sum()['total_runs']
```

### Add  +1  in all first inning runs for total runs which is target

```python
total_score_df = df.groupby(['match_id','innings']).sum()['total_runs'].reset_index()
total_score_df = total_score_df[total_score_df['innings'] == 1]
total_score_df["total_runs"] = total_score_df["total_runs"] + 1
```

```python
total_score_df
```

### New column for total runs which is target

```python
df = df.merge(total_score_df[['match_id','total_runs']],left_on='match_id',right_on='match_id')
```

```python
df
```

```python
df['batting_team'].unique()
```

### 'Kochi Tuskers Kerala', 'Rising Pune Supergiants' and 'Pune Warriors' are old teams and not playing in IPL 2024

```python
teams = [
    'Gujarat Titans',
    'Mumbai Indians',
    'Chennai Super Kings',
    'Lucknow Super Giants'
    'Rajasthan Royals',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Sunrisers Hyderabad',
    'Punjab Kings',
    'Delhi Capitals'
]
```

### Same teams with two diffrent names

```python
df['batting_team'] = df['batting_team'].str.replace('Gujarat Lions','Gujarat Titans')
df['bowling_team'] = df['bowling_team'].str.replace('Gujarat Lions','Gujarat Titans')

df['batting_team'] = df['batting_team'].str.replace('Delhi Daredevils','Delhi Capitals')
df['bowling_team'] = df['bowling_team'].str.replace('Delhi Daredevils','Delhi Capitals')

df['batting_team'] = df['batting_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
df['bowling_team'] = df['bowling_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

df['batting_team'] = df['batting_team'].str.replace('Kings XI Punjab','Punjab Kings')
df['bowling_team'] = df['bowling_team'].str.replace('Kings XI Punjab','Punjab Kings')
```

```python
df = df[df['batting_team'].isin(teams)]
df = df[df['bowling_team'].isin(teams)]
```

```python
df
```

### Rain interrupted matches

```python
df['dl_applied'].value_counts()
```

### Only second innings to chase the runs

```python
df = df[df['innings'] == 2]
```

### Cumulative runs of each second inning

```python
df.groupby('match_id')['total_runs_x'].cumsum()
```

### Current score of each ball

```python
df['current_score'] = df.groupby('match_id')['total_runs_x'].cumsum()
```

### Runs left to chase and win the match

```python
df['runs_left'] = df['total_runs_y'] - df['current_score']
```

```python
df.columns
```

### Balls left

```python
df['balls_left'] = 120 - (df['over']*6 + df['ball.1'])
```

```python
df
```

### Cumulative wickets of each second inning

```python
df['player_dismissed'] = df['player_dismissed'].fillna("0")
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
df['player_dismissed'] = df['player_dismissed'].astype('int')

wickets = df.groupby('match_id')['player_dismissed'].cumsum()
df['wickets'] = 10 - wickets
df.head()
```

### Current run rate = runs/overs

```python
df['crr'] = (df['current_score']*6)/(120 - df['balls_left'])
```

### Required run rate = runs left/balls left

```python
df['rrr'] = (df['runs_left']*6)/df['balls_left']
```

### Batting team wins or loses

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

### New column of batting team result

```python
df['result'] = df.apply(result,axis=1)
```

### Final data frame

```python
final_df = df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_y','crr','rrr','result']]
```

```python
final_df
```

### Shuffling for better accuracy

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df
```

### Drop null values

```python
final_df.dropna(inplace=True)
```

### Neglecting 0 ball left value to prevent infinite rrr

```python
final_df = final_df[final_df['balls_left'] != 0]
```

### Train Test Split

```python
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

### ColumnTransformer

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

### Train the model

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

### Accuracy of the model

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
